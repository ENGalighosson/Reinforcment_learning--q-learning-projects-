```python
import gym
import numpy as np

env = gym.make("FrozenLake-v1", is_slippery=False)

LEARNING_RATE = 0.1
DISCOUNT = 0.95
EPISODES = 10000

epsilon = 1.0
epsilon_decay = 0.0001
```

```python
frozenlake_q_table = np.zeros((env.observation_space.n, env.action_space.n))

print(frozenlake_q_table)
print(frozenlake_q_table.shape)
```

```python
rewards_per_episode = []

for episode in range(EPISODES):
    state, info = env.reset()
    done = False
    total_reward = 0

    while not done:
        if np.random.random() < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(frozenlake_q_table[state])

        new_state, reward, terminated, truncated, info = env.step(action)
        done = terminated or truncated

        current_q = frozenlake_q_table[state, action]
        max_future_q = np.max(frozenlake_q_table[new_state])

        new_q = (1 - LEARNING_RATE) * current_q + LEARNING_RATE * (
            reward + DISCOUNT * max_future_q
        )

        frozenlake_q_table[state, action] = new_q

        state = new_state
        total_reward += reward

    epsilon = max(epsilon - epsilon_decay, 0)
    rewards_per_episode.append(total_reward)

    

env.close()

print("\nFinal FrozenLake Q-table:")
print(frozenlake_q_table)
```
